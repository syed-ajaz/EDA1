```python
def stream(text):
    if text == 'computer engineering' or text == 'computer science & engineering' or text == 'computer science and technology' or text == 'computer and communication engineering' or text == 'computer science':
        return 'computer science engineering'
    else:
        return text
```

```python
df['Specialization'] = df['Specialization'].apply(stream)
```

```python
sp = df['Specialization'].value_counts()
```

```python
sp_df = sp[sp <= 15]
```

```python
def sp_combine(value):
    if value in sp_df:
        return "others"
    else:
        return value
```
